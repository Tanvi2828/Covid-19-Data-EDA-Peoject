```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# To display plots inline in Jupyter notebook
%matplotlib inline
```

```python
# Load dataset
df = pd.read_csv('country_wise_latest.csv')

# Check the first few rows
df.head()
```

```python
# Get data info
df.info()

# Check for missing values
df.isnull().sum()
```

```python
# Dropping rows with missing values (you can also choose to fill with mean/median)
df = df.dropna()
```

```python
# Check the column names
df.columns
```

```python
# Check the first few rows of the dataset
df.head()

# Get general information about the dataset (data types, missing values)
df.info()

# Summary statistics (mean, std, min, max) for numerical columns
df.describe()
```

```python
# Check for missing values
df.isnull().sum()

# You can fill or drop missing values (depending on the analysis)
# Example: Fill missing values with 0 for numerical columns
df.fillna(0, inplace=True)
```

```python
# Get unique values in categorical columns
print(df['Country/Region'].unique())
print(df['WHO Region'].unique())

# Check data types
df.dtypes
```

```python
# Distribution of Confirmed, Deaths, and Recovered cases
import matplotlib.pyplot as plt
import seaborn as sns

# Plot histogram for numerical columns
df[['Confirmed', 'Deaths', 'Recovered', 'Active']].hist(bins=20, figsize=(10,6))
plt.suptitle('Distribution of Confirmed, Deaths, Recovered, Active Cases')
plt.show()

# Correlation heatmap to see how variables relate to each other
plt.figure(figsize=(10,6))
sns.heatmap(df[['Confirmed', 'Deaths', 'Recovered', 'Active']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Sort countries by confirmed cases and plot the top 10 countries
top_countries_confirmed = df.groupby('Country/Region')['Confirmed'].max().sort_values(ascending=False).head(10)
top_countries_deaths = df.groupby('Country/Region')['Deaths'].max().sort_values(ascending=False).head(10)
top_countries_recovered = df.groupby('Country/Region')['Recovered'].max().sort_values(ascending=False).head(10)

# Plot bar charts for top countries
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

top_countries_confirmed.plot(kind='bar', ax=axes[0], color='skyblue')
axes[0].set_title('Top 10 Countries by Confirmed Cases')
axes[0].set_ylabel('Confirmed Cases')
axes[0].set_xticklabels(top_countries_confirmed.index, rotation=45)

top_countries_deaths.plot(kind='bar', ax=axes[1], color='salmon')
axes[1].set_title('Top 10 Countries by Deaths')
axes[1].set_ylabel('Deaths')
axes[1].set_xticklabels(top_countries_deaths.index, rotation=45)

top_countries_recovered.plot(kind='bar', ax=axes[2], color='lightgreen')
axes[2].set_title('Top 10 Countries by Recovered Cases')
axes[2].set_ylabel('Recovered Cases')
axes[2].set_xticklabels(top_countries_recovered.index, rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Calculate Deaths per Case ratio
df['Deaths / 100 Cases'] = df['Deaths'] / df['Confirmed'] * 100

# Calculate Recovered per Case ratio
df['Recovered / 100 Cases'] = df['Recovered'] / df['Confirmed'] * 100

# Plot these ratios for the top 10 countries
fig, axes = plt.subplots(1, 2, figsize=(18, 6))

df.groupby('Country/Region')['Deaths / 100 Cases'].max().sort_values(ascending=False).head(10).plot(kind='bar', ax=axes[0], color='darkred')
axes[0].set_title('Top 10 Countries by Deaths / 100 Cases')
axes[0].set_ylabel('Deaths / 100 Cases')

df.groupby('Country/Region')['Recovered / 100 Cases'].max().sort_values(ascending=False).head(10).plot(kind='bar', ax=axes[1], color='darkgreen')
axes[1].set_title('Top 10 Countries by Recovered / 100 Cases')
axes[1].set_ylabel('Recovered / 100 Cases')

plt.tight_layout()
plt.show()
```

```python
# Save the plots to images
plt.figure(figsize=(12,6))
sns.barplot(x='Country/Region', y='1 week change', data=df.sort_values(by='1 week change', ascending=False).head(10))
plt.title('Top 10 Countries by 1 Week Change in Confirmed Cases')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('top_10_countries_by_week_change.png')

# Similarly, save other plots
```
